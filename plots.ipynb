```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("results.csv")
df['ArraySize'] = 2**df['N']
```

```python
plt.style.use('seaborn-v0_8-whitegrid')
fig, ax = plt.subplots(figsize=(12, 8))


ax.plot(df['ArraySize'], df['CPU'], marker='o', linestyle='--', label='CPU Bitonic Sort')
ax.plot(df['ArraySize'], df['GPU_Bitonic'], marker='s', linestyle='-', label='GPU Bitonic Sort (Total)')
ax.plot(df['ArraySize'], df['GPU_MergeSort'], marker='^', linestyle='-', label='GPU Merge Sort (Total)')


ax.set_xscale('log', base=2)
ax.set_yscale('log')


ax.set_title('Sorting Algorithm Performance: CPU_Bitonic vs. GPU_Bitonic vs GPU_Merge', fontsize=16, weight='bold')
ax.set_xlabel('Array Size (N)', fontsize=12)
ax.set_ylabel('Time (milliseconds)', fontsize=12)


ax.legend(fontsize=11)


ax.grid(True, which="both", ls="--", linewidth=0.5)

# Ensuring everything fits nicely
plt.tight_layout()


plt.savefig("sorting_performance_chart.png", dpi=300, bbox_inches='tight')


plt.show()

print("sorting_performance_chart.png")
```
